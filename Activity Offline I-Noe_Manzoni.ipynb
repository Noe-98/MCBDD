```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def interference (infected, p_healty):
    healty = 1 - infected
    p_falseNegative = 1 - p_healty
    p = ((0.99 * infected)/(0.99 * infected + p_falseNegative * healty))
    return p
    
```

```python
infected = np.arange(0.00001, 0.5 + 0.00001, 0.00001)
#p_healty = 0.99, 0.999, 0.99999

p = np.zeros(len(infected))
g = np.zeros(len(infected))
v = np.zeros(len(infected))
for i in range(len(infected)):
    p[i] = interference(infected[i], 0.99) * 100
    g[i] = interference(infected[i], 0.999) * 100
    v[i] = interference(infected[i], 0.99999) * 100
print(v)
plt.plot(infected*100, p, '--', label='specificity 0.99')
plt.plot(infected*100, g, '--', label='specificity 0.999')
plt.plot(infected*100, v, '--', label='specificity 0.99999')
plt.legend()
plt.xlabel('infected rate %')
plt.ylabel('probabilities of beeing indeed positive %')
plt.show()
```
