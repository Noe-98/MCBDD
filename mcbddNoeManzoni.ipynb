#### First part of analysis

This code gave us the possibility to explore the database a little bit since we have not found any class or relation ship diagrams online. We used it to reason about where the data can be found and its possible relations between the exisiting 57 tables.

```python
%sql SELECT * FROM information_schema.columns WHERE table_schema = 'public' ORDER BY column_name
```

With the following code-line we make sure that our drug is found in the database

```python
%sql SELECT * FROM product WHERE product_name = 'Donepezil'
```

The following two queries extract the id/struct_id of the drug to use for further queries

```python
%sql SELECT * FROM synonyms WHERE lname LIKE 'donepezil%' LIMIT 50
```

```python
%sql SELECT DISTINCT id FROM synonyms WHERE lname LIKE 'donepezil%' LIMIT 10
```

After we have found the id for Donepezil, which is '946', we are able to extract its target and the target type.

```python
%sql SELECT target_class, COUNT(target_class) FROM act_table_full WHERE struct_id = 946 GROUP BY target_class
```

#### Second part of analysis

Looking for a drug which targets less different target types by trial and error (we used the web GUI of drugcentral.org and searched there with the keyword "Alzheimer". With the result generated by the search, we have found that "Galantamine" to have less target types compared to "Donepezil". The queries for this are the same as in the first part:

```python
%sql SELECT DISTINCT id FROM synonyms WHERE lname LIKE 'galantamine%'
```

Pasted the found id from the query before into the next query:

```python
%sql SELECT target_class, COUNT(target_class) FROM act_table_full WHERE struct_id = 1272 GROUP BY target_class
```

We now compare those two different drugs of its reported sideffects using the table. We can observe that there are far less side effects reported for the first query which represents the drug "Galantamine" (row count: 237 for Galantamine vs. 791, and highest likelihood level (llr, 6th column) for "Loss of consciousness" for example: likelihood level 80.62 for Galantamine vs. 364.
Note that we do  not exactly figured which value or ratio is represented exactly with the likelihood ratio numbers provided, nevertheless, the number of reported side effects may state, that there are more side effects found with a drug that has more target types as a target. Also note that this is only an assumption and we only found one case which "fullfills" our hypothesis. For further analysis we would look for more confirmations and or negations to better reasono about our hypothesis.

```python
%sql SELECT * FROM faers WHERE struct_id = 1272 ORDER BY llr DESC
```

```python
%sql SELECT * FROM faers WHERE struct_id = 946 ORDER BY llr DESC
```
