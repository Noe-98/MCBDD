#### First part of analysis

This code gave us the possibility to explore the database a little bit since I have not found any class or relation ship diagrams online. I used it to reason about where the data can be found and its possible relations between the exisiting 57 tables.

```python
%sql SELECT * FROM information_schema.columns WHERE table_schema = 'public' ORDER BY column_name
```

With the following code-line I make sure that our drug is found in the database

```python
%sql SELECT * FROM product WHERE product_name = 'Chlorothiazide'
```

The following two queries extract the id/struct_id of the drug to use for further queries

```python
%sql SELECT * FROM synonyms WHERE lname LIKE 'chlorothiazide%'
```

```python
%sql SELECT DISTINCT id FROM synonyms WHERE lname LIKE 'chlorothiazide%' 
```

After I have found the id for Chlorothiazide, which is '609', we are able to extract its target and the target type.

```python
%sql SELECT target_class, COUNT(target_class) FROM act_table_full WHERE struct_id = 609 GROUP BY target_class
```
